## Part 4: ORDER BY - Sorting (4 queries)

### Query 15: Sort by Age (Youngest First)

**Task:** Show Name and Age, sorted by age from youngest to oldest. Limit to 10.

### Query 16: Sort by Age (Oldest First)

**Task:** Show Name and Age, sorted by age from oldest to youngest. Limit to 10.

**Hint:** Use DESC!

```python
# TODO: Write your query
query_16 = """
SELECT Name, Age
FROM passengers
ORDER BY age DESC
LIMIT 10
"""

result_16 = pd.read_sql(query_16, conn)
display(result_16)
```

## Part 5: Combined Queries (3 challenges)

Put it all together: SELECT, WHERE, ORDER BY, LIMIT

### Query 19: Young Female Survivors

**Task:** Find female passengers under age 20 who survived. Show Name, Age, Sex, Survived. Sort by age (youngest first). Limit to 10.

```python
# TODO: Write your query
query_19 = """
SELECT Name, Age, Sex, Survived
FROM passengers
WHERE sex = 'female' AND age < 20 AND survived = 1
ORDER BY age ASC
LIMIT 10
"""

result_19 = pd.read_sql(query_19, conn)
display(result_19)
```

### Query 20: Wealthy First Class Passengers

**Task:** Find first class passengers who paid more than $100. Show Name, Pclass, Fare. Sort by fare (highest first). Limit to 10.
