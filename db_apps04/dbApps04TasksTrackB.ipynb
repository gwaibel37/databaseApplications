```python
import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect('titanic.db')
print("✅ Connected to database")
```

```python
query = """
    SELECT Name, Age, Sex, Pclass, Survived, SibSp, Fare
    FROM passengers
    WHERE Sex = 'male' AND Age >=16 AND Age <= 18
        AND Pclass = 3 AND Survived = 1
    ORDER BY Fare ASC
    LIMIT 10;
"""

result = pd.read_sql(query, conn)
display(result)
```
